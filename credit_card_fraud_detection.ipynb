```python
import matplotlib.pyplot as plt

def pie_bar_graphs(df):
  fraud_counts = df.Class.value_counts()
  labels = ['Normal transaction','Fraud']
  #graph
  fig, (ax1,ax2) = plt.subplots(ncols=2,figsize=(15,8))
  #pie chart
  ax1.pie(fraud_counts,labels=labels,colors=['green','red'],autopct='%1.1f%%')
  ax1.set_title("Transaction happened or not ")

  #bar plot
  ax2.bar(labels,fraud_counts.values,color=['green','red'])
  ax2.set_xlabel("Transaction Class")
  ax2.set_ylabel("Number of transactions")
  ax2.set_title("Credit card fraud detection")
  for i in ax2.containers:
    ax2.bar_label(i,)

  plt.show()

```

```python
pie_bar_graphs(df)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

def box_plots(df, columns_list, rows, cols, suptitle):
    fig, axs = plt.subplots(rows, cols, sharey=True, figsize=(14,20))
    axs = axs.flatten()

    for i, col in enumerate(columns_list):
        sns.boxplot(x=df[col], orient='h', ax=axs[i])

        skew = round(df[col].skew(), 2)
        axs[i].set_title(f"{col}, skewness: {skew}")

    fig.suptitle(suptitle, fontsize=16)
    plt.tight_layout()
    plt.show()
box_plots(df=df, columns_list=numeric_columns, rows=10, cols=3, suptitle='')
plt.tight_layout()
```

# Seprate training and test data

```python
X = df.drop('Class',axis=1)
y=df['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

# when ImBalanced dataset

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, roc_auc_score, average_precision_score

xgb_baseline = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    eval_metric='aucpr',
    random_state=42
)

xgb_baseline.fit(X_train, y_train)


y_pred_baseline = xgb_baseline.predict(X_test)
print("Baseline Model:\n", classification_report(y_test, y_pred_baseline))
```

#after applying SMOTE to create a synthetic data

```python
import numpy as np
from sklearn.metrics import (classification_report, roc_auc_score,
                             average_precision_score, precision_recall_curve)
from imblearn.over_sampling import SMOTE

# Apply SMOTE only on training data (NEVER on test!)
sm = SMOTE(random_state=42)
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)

print(f"Before SMOTE: {dict(y_train.value_counts())}")
print(f"After  SMOTE: {dict(pd.Series(y_train_sm).value_counts())}")
```

```python
from xgboost import XGBClassifier


xgb = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    eval_metric='aucpr',     # PR-AUC better than logloss for imbalance
    random_state=42
)

xgb.fit(
    X_train_sm, y_train_sm,
    eval_set=[(X_test, y_test)],
    verbose=50
)

y_proba_xgb = xgb.predict_proba(X_test)[:, 1]
y_predict_xgb = xgb.predict(X_test)
print("XGBoost PR-AUC:", average_precision_score(y_test, y_proba_xgb))
print("XGBoost Model:\n", classification_report(y_test, y_predict_xgb))
```

## SOME FINE TUNNING IN XGBoost. I tried it in n_estimators and learning_rate

```python
from xgboost import XGBClassifier


xgb = XGBClassifier(
    n_estimators=300,
    max_depth=7,
    learning_rate=0.07,
    eval_metric='aucpr',     # PR-AUC better than logloss for imbalance
    random_state=42
)

xgb.fit(
    X_train_sm, y_train_sm,
    eval_set=[(X_test, y_test)],
    verbose=50
)

y_proba_xgb = xgb.predict_proba(X_test)[:, 1]
y_predict_xgb = xgb.predict(X_test)

print("XGBoost PR-AUC:", average_precision_score(y_test, y_proba_xgb))
print("XGBoost Model:\n", classification_report(y_test, y_predict_xgb))
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

precision, recall, thresholds = precision_recall_curve(y_test, y_proba_xgb)

plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.show()
```
